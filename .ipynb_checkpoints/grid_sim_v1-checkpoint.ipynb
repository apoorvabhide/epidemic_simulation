```python
# Now, including dynamic 
# Create a small grid
X = 100
Y = 100
N = X*Y
beta = 0.033 #Parameter for contagiousness
p = 0.003
h = 0.5*math.log(p/(1-p))
inf_recovery_time = 11
reinf_immunity_time = 90
rate = 0.00144
```

```python
# Initiating a very basic agent: has a position and Status (0 = Susceptible, 1 = Infected).
# Also stores the positions of its neighbours, to make it easy to check infected status etc.
#cvals  = [0,1]
#colors = ["green","red"]

#norm=plt.Normalize(min(cvals),max(cvals))
#tuples = list(zip(map(norm,cvals), colors))
#cmap = ListedColormap()
cmap = ListedColormap(['green', 'red'])
norm = BoundaryNorm(boundaries = [-0.05,0.5,1.5],ncolors=cmap.N)

class Agent:
    def __init__(self,x, y,network=None):
        self.Location = [x,y]
        self.Network = network
        self.Status = 0
        self.t_inf = 0
        self.t_rec = 0
        self.Recovery = np.random.normal(loc = 13, scale = 6)
        self.Neighbours = [[x-1,y],[x,y-1],[x,y+1],[x+1,y]]
        self.Neighbours = [item for item in self.Neighbours if item[0] != -1 and item[1] != -1 and item[0] < X and item[1] < Y]
        self.Contacts = []
        self.NNeighbours = round(np.random.exponential(scale = 5))
        if self.NNeighbours > 50:
            self.NNeighbours = 50
            
        self.Appetite = self.NNeighbours
        
    def __setitem__(self,value):
        self.Status = value
        
    def NeighbourAdd(self,new_neighbour):
        self.Neighbours.append(new_neighbour)
        
    def situate(self,network_name):
        self.Network = network_name
        self.InfectedNeighbours = [agents[self.Neighbours[i]].Status for i in range(len(self.Neighbours))].count(1)
        self.InfectionProbability = math.exp(4*beta*self.InfectedNeighbours + 2*h)/(1 + math.exp(4*beta*self.InfectedNeighbours + 2*h))
        
    def update(self,config):
        
        # Config 1 has no immunity. On the 11th day people become susceptible again.
        if config == 1:
            self.InfectedNeighbours = [agents[self.Neighbours[i]].Status for i in range(len(self.Neighbours))].count(1)
            self.InfectionProbability = math.exp(4*beta*self.InfectedNeighbours + 2*h)/(1 + math.exp(4*beta*self.InfectedNeighbours + 2*h))
            if self.Status == 0:
                if random.random() < self.InfectionProbability:
                    self.Status = 1
            elif self.Status == 1:
                self.t_inf += 1
                if self.t_inf > inf_recovery_time:
                    self.Status = 0
                    self.t_inf = 0
 
        # Config 2 means permanent immunity: on the (tau+1) day, people become immune forever.
        elif config == 2:
            self.InfectedNeighbours = [agents[self.Neighbours[i]].Status for i in range(len(self.Neighbours))].count(1)
            self.InfectionProbability = math.exp(4*beta*self.InfectedNeighbours + 2*h)/(1 + math.exp(4*beta*self.InfectedNeighbours + 2*h))
            if self.Status == 0:
                if random.random() < self.InfectionProbability:
                    self.Status = 1
            elif self.Status == 1:
                self.t_inf += 1
                if self.t_inf >= inf_recovery_time:
                    self.Status = 2
                    self.t_inf = 0
            elif self.Status == 2:
                self.t_rec += 1
        
        # Config 3 means time-limited immunity: immunity lasts until day 90 of being recovered.
        # Then this immunity either goes completely away or goes away with 50% probability.
        elif config == 3:
            self.InfectedNeighbours = [agents[self.Neighbours[i]].Status for i in range(len(self.Neighbours))].count(1)
            self.InfectionProbability = math.exp(4*beta*self.InfectedNeighbours + 2*h)/(1 + math.exp(4*beta*self.InfectedNeighbours + 2*h))
            if self.Status == 0:
                if random.random() < self.InfectionProbability:
                    self.Status = 1
            elif self.Status == 1:
                self.t_inf += 1
                if self.t_inf >= inf_recovery_time:
                    self.Status = 2
                    self.t_inf = 0
            elif self.Status == 2:
                self.t_rec += 1
                if self.t_rec == reinf_immunity_time:
                    if random.random() < 0.5:
                        self.Status = 3
                        self.t_inf = 0
                    else:
                        self.Status = 0
                        self.t_inf = 0
                        self.t_rec = 0

        # Config 4 means exponentially decaying immunity: Immunity keeps decreasing from day of recovery,
        # with a median time to reinfection of 16 months. Additionally, recovery is also stochastic with a
        # normally distributed duration of recovery ~ N(13,6).
        elif config == 4:
            self.InfectedNeighbours = [agents[self.Neighbours[i]].Status for i in range(len(self.Neighbours))].count(1)
            self.InfectionProbability = math.exp(4*beta*self.InfectedNeighbours + 2*h)/(1 + math.exp(4*beta*self.InfectedNeighbours + 2*h))
            if self.Status == 0:
                if random.random() < self.InfectionProbability:
                    self.Status = 1
            elif self.Status == 1:
                self.t_inf += 1
                if self.t_inf >= self.Recovery:
                    self.Status = 2
                    self.t_inf = 0
            elif self.Status == 2:
                self.ReinfRisk = 1 - math.exp(-imm_decay_rate*self.t_rec)
                self.t_rec += 1
                # Agent can become susceptible again with given exponentially decaying probability
                if random.random() < self.ReinfRisk:
                    self.Status = 0
        
        # Config 5 means exponentially decaying immunity: Immunity keeps decreasing from day of recovery,
        # with a median time to reinfection of 16 months. Additionally, recovery is also stochastic with a
        # normally distributed duration of recovery ~ N(13,6).
        elif config == 5:
            self.InfectedNeighbours = [agents[self.Contacts[i]].Status for i in range(len(self.Contacts))].count(1)
            #self.InfectionProbability = math.exp(4*beta*self.InfectedNeighbours + 2*h)/(1 + math.exp(4*beta*self.InfectedNeighbours + 2*h))
            if self.InfectedNeighbours == 0:
                self.InfectionProbability = p
            else:
                self.InfectionProbability = beta*math.log(self.InfectedNeighbours) + p
            if self.Status == 0:
                #print(self.InfectionProbability)
                if random.random() < self.InfectionProbability:
                    self.Status = 1
            elif self.Status == 1:
                self.t_inf += 1
                if self.t_inf >= self.Recovery:
                    self.Status = 2
                    self.t_inf = 0
            elif self.Status == 2:
                self.ReinfRisk = 1 - math.exp(-imm_decay_rate*self.t_rec)
                self.t_rec += 1
                # Agent can become susceptible again with given exponentially decaying probability
                if random.random() < self.ReinfRisk:
                    self.Status = 0
            
            self.NNeighbours = round(np.random.gamma(shape = 30,scale = 0.1643))
            self.Appetite = self.NNeighbours
                    
        

class AgentSet(Agent):
    def __init__(self,X, Y):
        self.agents = [[Agent(x,y) for x in range(X)] for y in range(Y)]
    
    def __len__(self):
        return(X*Y)
    
    def print_agents(self):
        for x in range(X):
            for y in range(Y):
                print(self.agents[x][y].Location)
                
    def print_contacts(self):
        for x in range(X):
            for y in range(Y):
                print(self.agents[x][y].Contacts)
                    
    def plot_grid(self):
        df = pd.DataFrame(columns=['x', 'y', 'Status'])
        i = -1
        for x in range(X):
            for y in range(Y):
                i +=1
                df.loc[i] = [x] + [y] + [agents[x,y].Status]
        df = df.pivot_table(index=['x'], columns=['y'], values=['Status'])
        fig,ax = plt.subplots()
        hm = sns.heatmap(data=df,cmap=cmap,norm=norm,ax=ax)
        hm.set(xticklabels=[])
        hm.set(yticklabels=[])
        hm.tick_params(bottom=False)  # remove the ticks
        hm.tick_params(left=False)  # remove the ticks
        hm.set(xlabel=None)
        hm.set(ylabel=None)
        #cbar = plt.colorbar(hm)
        #cbar.set_yticklabels(['Susceptible','Infected'])
        #cax = divider.append_axes('bottom', size='10%', pad=0.6)
        #cb = fig.colorbar(heatmap, cax=cax, orientation='horizontal')
        #cbar.set_label('# of contacts', rotation=270)

    def __getitem__(self,pos):
        x_id,y_id=pos
        #print("fetching %s,%s" % (x_id,y_id))
        return(self.agents[x_id][y_id])
    
    def __setitem__(self,pos,value):
        x_id,y_id=pos
        self.agents[x_id][y_id].Status = value
        
    def update_grid(self,config):
        for x in range(X):
            for y in range(Y):
                self.agents[x][y].update(config)
                
    def summarise(self):
        n_infected = [agents[x,y].Status for x in range(X) for y in range(Y)].count(1)
        n_recovered = [agents[x,y].Status for x in range(X) for y in range(Y)].count(2)
        self.InfectedPercent = n_infected/(X*Y)
        self.RecoveredPercent = n_recovered/(X*Y)
        print("Percentage of Infected population: %s" % self.InfectedPercent)
        print("Percentage of Recovered population: %s" % self.RecoveredPercent)
        
    def assign_contacts(self):
        for x in range(X):
            for y in range(Y):
                # Idea: for a particular agent on the grid, sample agents around it using a bivariate
                # normal density. If this agent has a non-zero appetite (ie hasn't used up all its
                # contacts already), it is now the present agent's neighbour and vice versa.
                #print("Point being assigned: ",[x,y])
                mean = [x,y]
                cov = [[5,0],[0,5]]
                #print("Appetite: ",agents[x,y].Appetite)
                contacts = []
                n=0
                excl =0
                # Insert while loop here for generating contacts
                while agents[x,y].Appetite > 0:
#                for i in range(10):
                    contact = np.rint(np.random.multivariate_normal(mean=mean,cov=cov)).tolist()
                    if(not all(j >= 0 for j in contact) or not all(j < X and j < Y for j in contact)):
                        #print('excl')
                        excl +=1
                        if excl > 100:
                            #print("Skipping...")
                            break
                        continue
                    if(agents[list(map(int,contact))].Appetite == 0 or [x,y] == list(map(int,contact))):
                        #print("excl")
                        excl +=1
                        if excl > 100:
                            #print("Skipping...")
                            break
                        continue
                    contact = list(map(int,contact))
                    # Add the current agent to this new agent's Contacts, decrease both guys' appetites.
                    agents[contact].Contacts.append([x,y])
                    agents[x,y].Contacts.append(contact)
                    agents[x,y].Appetite -= 1
                    agents[contact].Appetite -= 1
                    #print(agents[x,y].Contacts)
                    n+=1
```

```python
# Now, we simulate an outbreak from the beginning

# Scenario 1: SIS model, where on day 11 everyone goes back to being susceptible. In this scenario we get
# almost a Malthusian curve, asymptotic around a peak value determined by beta and h. Due to the parameters 
# chosen from the chinese study, our virus is a lot more slow-moving than the 

# 1. Initialise grid
nsim = 50
for sim in range(nsim):
    agents = AgentSet(X = X,Y=Y)
    for x in range(X):
        for y in range(Y):
            agents[x,y].situate(network_name=agents)

    df = pd.DataFrame(columns=['SimNo','t', 'Infected','Recovered'])        
    T = 100 # 100 time steps to begin with
    for t in range(T):
        print("Simulation No: ",sim,"Day %s" % t)
        agents.update_grid(config=1)
        agents.summarise()
        inf = agents.InfectedPercent
        rec = agents.RecoveredPercent
        df.loc[t] = [nsim] + [t] + [inf] + [rec]
```

We know that immunity isn't a hard bound like that - it keeps decaying over time, and doesn't just disappear one day. To capture this, we use parameters from [this paper](https://www.thelancet.com/journals/lanmic/article/PIIS2666-5247(21)00219-6/fulltext). It suggests that the probability of not being reinfected is 1 on Day 1 immediately after recovery, and then it decreases exponentially at a constant rate with a median time to reinfection of 16 months. Consequently, we use a decaying exponential function that starts at 1 and decays with a rate parameter of 0.00144 (fitting the median of 16 months ~ 480 days).

Further, a [study from Ethiopia] (https://www.ncbi.nlm.nih.gov/pmc/articles/PMC8673952/) found that the time to recovery from COVID was roughly symmetrically distributed, with a median of 13 days and an IQR of 9-17. Here, we assume this to be Gaussian, which would mean $d \sim N (13, 6) $.

```python
# Scenario 4: SIR model with decaying immunity and stochastic recovery. 
# In this case, Recovery isn't just a 10 day thing - number of days to recovery are normally distributed with a
# mean of 13 and an sd of 6.

# We continue with tau = 10 and exponentially decaying immunity
# after 90 days.
X = 250
Y = 250
N = X*Y
beta = 0.48 #Parameter for contagiousness
p = 0.033
h = 0.5*math.log(p/(1-p))
inf_recovery_time = 10
reinf_immunity_time = 90
rate = 0.00144

# 1. Initialise grid
agents = AgentSet(X = X,Y=Y)
for x in range(X):
    for y in range(Y):
        agents[x,y].situate(network_name=agents)
        
df = pd.DataFrame(columns=['t', 'Infected','Recovered'])        
T = 400 # Now we simulate a longer duration to see the effect of immunity
for t in range(T):
    print("Day %s" % t)
    agents.update_grid(config=4)
    agents.summarise()
    inf = agents.InfectedPercent
    rec = agents.RecoveredPercent
    df.loc[t] = [t] + [inf] + [rec]
```

[This paper](https://www.medrxiv.org/content/10.1101/2020.03.24.20042606v1.full-text) states that the probability of getting COVID from a "close/household" contact was 10.2%, and the probability of getting COVID from a random contact outside the social network (like public transport/healthcare) was ~0.3%. In our model, we assume that the close/household contact is with someone the agent is connected to in their social network, and an outside contact is otherwise. Using these parameters, we get a $\beta$ = 0.36 and $h = -1.88$. 

Furthermore, the CoMix paper suggests a lack of uniformity in the number of contacts. The number of contacts is said to have a right-skewed distribution with a long right tail - some individuals report as high as 50 contacts in a day, but the mean number of contacts is ~5 in periods with eased restrictions (and only 13% of individuals had >6 contacts, as shown [here](https://assets.publishing.service.gov.uk/government/uploads/system/uploads/attachment_data/file/1009157/S1281_CoMix_Report_-_high_numbers_of_contacts_SAGE_91.pdf).

Clearly, this isn't adequately represented by a lattice where every individual has four contacts in a given day. We model this using an exponential distribution with a mean of 5 (this is a beginning, other choices like negative binomial distribution are also viable).
